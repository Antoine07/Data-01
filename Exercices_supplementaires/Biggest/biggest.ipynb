```python

import numpy as np

def is_order_numbers(n):

    numbers = np.array( [int(num) for num in str(n)], dtype=int )
    if len(numbers) == 1:
        return True

    for i in range( len(numbers) - 1 ):
        if numbers[i] >= numbers[i+1]:
            return False

    return True


# On doit en trouver 511
count = 10
list_order_numbers = np.array([], dtype=int)

while count < 123456789:
    if is_order_numbers(count):
        list_order_numbers = np.append( list_order_numbers,  count ) 
    count += 1


print(list_order_numbers )




```
