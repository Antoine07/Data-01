```python
import pandas as pd

notes = pd.read_csv('./data/notes_statistiques.csv')

notes
```

```python
notes['Frequence'] = notes[['Effectifs']] / notes['Effectifs'].sum()
notes = notes.rename(columns={'Notes examen de statistiques': 'Notes'})

notes.head()
```

```python
%matplotlib notebook
import matplotlib.pyplot as plt

fig = plt.figure()

width = 0.05

plt.bar(notes['Notes'].values, notes['Effectifs'].values, width, color='b' )

plt.show()
```

```python
notes = notes.rename(columns = {'Notes examen de statistiques' : 'Notes'})

notes
```

```python

notes['notes_bin'] = pd.qantiles(x=notes['Notes'], bins=[0, 4, 8, 10, 12, 14, 18, 21])
print(notes)

grouped = notes.groupby('notes_bin')

dfNotes = grouped[['Effectifs']].sum()

dfNotes['Frequences'] = dfNotes['Effectifs'] / dfNotes['Effectifs'].sum()

dfNotes.round(2)

```

```python
# Fermeture de l'objet matplotlib
plt.close()
```

```python
%matplotlib notebook
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="ticks", color_codes=True)

sns.catplot(x="Effectifs", y ='Frequences', kind="bar", palette="ch:.25", data=dfNotes);
```
