## while

la structure du while est la suivante :

le mot réservé while et une condition boolean True ou Flase. Si la condition est vrai on exécute les instructions dans la boucle while et si la condition est fausse on n'exécute pas les instructions dans la boucle

while condition:
    instruction1
    instruction2
    ...

```python

count = 0

# condition toujours vrai => donc on exécute le code dans le while
while True:
    # les instructions
    print("bonjour")

count = 0
while count < 10:
    count = count + 1
    print(count)
    
"""
# On faisant tourner le dernier code
count = 0

while v(0 < 10) = Vrai:
    count = 1
    affiche 1

while v(1 < 10) = Vrai:
    count = 2
    affiche 2
    
while v(2 < 10) = vrai :
    count = 3
    affiche 3

...

while v(9< 10) = vrai :
    count = 10
    affiche 10
    
while v(10 < 10) = faux => la boucle s'arrête
"""
```

````python
count = 0
# tant que la condition est vraie on exécute les instructions dans la boucle
# si la condition est fausse alors on ne rentre pas dans la boucle
while count < 3 :
    count = count + 1
    print("bonjour", count)
    
print("terminé")


# Vous pouvez également arrêter une boucle avec le mot réservé break dans la boucle
number = 8
count2 = 0

while True:
    count2 = count2 + 1
    
    print(count2)
    # le mot clé break permet de sortir de la boucle
    if number == count2:
        break
````

## 7. Exercice
Reprendre l'exercice 6 et écrire maintenant une fonction **estPremier**, elle prend un paramètre n entier et retourne True si le nombre est premier et False sinon.

````python
def estPremier(n):
    diviseur = 1
    # compter le nombre de diviseurs
    count = 0

    # on teste tous les diviseurs de 1 à nombre
    while diviseur <= n:

        if n % diviseur == 0:
            # si on trouve un diviseur on fait +1
            count = count + 1
        diviseur = diviseur + 1
        
    if count == 2:
        return True
    else:
        return False

p = 11
if estPremier(p):
    print("ce nombre est premier", p)
else:
    print("ce nombre n'est pas premier", p)
````

### Structure de la boucle for

Une boucle for peut se traduire comme suit :

```python
"""

pour i de 0 à 9 faire :
    instructions

"""
for i in range(10):
    # instructions ...
    
```


range est une fonction Python qui génère une liste de  0 à 10. Cette fonction est intéressante car vous pouvez agir sur trois paramètres, un est obligatoire.

- range(11) 0 ... 10, la dernière valeur est la borne max non comprise

- range(0, 11)   0 ... 10

- range(1, 11)   1 ... 10 

- range(2, 11)   2 ... 10


- range(0, 100, 2)  0 ... 100 que les nombres pairs

- range(start, borne_sup, step )


La boucle for permet également d'itérer sur des listes en Python. Une liste est une structure de données permettant d'avoir une collection de valeurs dans une seule et même variable.

Une liste se définie avec des crochets ouvrant \[ et fermant c \]

```python

# les diviseurs de 12 rangés dans une liste
diviseurs = [ 1, 2, 3, 4, 6, 12 ]

# on peut afficher les diviseurs de 12
for num in diviseurs:
    print(num)

```

Une liste est une super variable qui contient des variables (variable de variables). Chaque variable de la liste est rangée dans la liste et possède un indice permettant d'accéder à la valeur de la liste.

```python

diviseurs = [ 1, 2, 3, 4, 6, 12 ]

# Pour accéder aux valeurs de la liste on utilise leur indices

diviseurs[0]
diviseurs[1]
diviseurs[2]
diviseurs[3]
diviseurs[4]
diviseurs[5]

```

attention l'indice 6 n'existe pas ici par exemple dans la liste précédente. Vous aurez un message d'erreur si vous essayez d'accéder à un indice de liste qui n'existe pas, voyez le type de message ci-dessous pour ce cas :

```python
# IndexError: list index out of range

```
