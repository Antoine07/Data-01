```python
%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt

# d√©finiton d'un style de type grille sous fond blanc
sns.set(style="whitegrid")
```

```python
fig, ax = plt.subplots(2, 3, sharex='col', sharey='row')
```

```python
tips = sns.load_dataset("tips")

tips.head()
```

```python
fig_dims = (10, 10)
fig, ax1 = plt.subplots(figsize=fig_dims)

sns.boxplot(x="total_bill", ax=ax1, data= tips)
plt.show()
```

```python
fig_dims = (20, 15)
fig, ax2 = plt.subplots(figsize=fig_dims)

sns.boxplot(x="day", y="total_bill", ax=ax2, data=tips)
plt.show()
```

```python
fig_dims = (20, 10)
fig, ax3 = plt.subplots(figsize=fig_dims)

sns.boxplot(x="day", y="total_bill", hue="smoker", ax=ax3,
                 data=tips, palette="Set3")

plt.show()
```
